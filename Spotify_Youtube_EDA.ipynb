```python
# Views Aykırı değerleri temizleme

def remove_outliers_views(df, column):
   
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
 
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
     
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df_cleaned_views = remove_outliers_views(df, 'Views')

print(df_cleaned_views.describe())
```

```python
# Likes Aykırı değerleri temizleme 

def remove_outliers_likes(df, column):
 
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
      
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
 
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df_cleaned_likes = remove_outliers_likes(df, 'Likes')

print(df_cleaned_likes.describe())
```

```python
#Comments aykırı değerleri temizleme: 

def remove_outliers_comments(df, column):
    
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
  
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df_cleaned_comments = remove_outliers_comments(df, 'Comments')


print(df_cleaned_comments.describe())
```

```python
# Sanatçıları ve bu puanların bir araya getirilip yüzdelenmesi
sanatci_enerji = df.groupby('Artist')['Energy'].mean()*100
sanatci_dans = df.groupby('Artist')['Danceability'].mean()*100
sanatci_duygusal = df.groupby('Artist')['Valence'].mean()*100

# Sanatçıların ortalama puanlarının birleşimi
sanatci_puanlari = pd.concat([sanatci_enerji, sanatci_dans, sanatci_duygusal], axis=1)

# Her bir özelliği (enerji, dans edilebilirlik, duygusallık) ayrı ayrı sıralanması
en_enerjik_sanatci = sanatci_puanlari.sort_values(by='Energy', ascending=False).head(1)
en_dans_edilebilir_sanatci = sanatci_puanlari.sort_values(by='Danceability', ascending=False).head(1)
en_duygusal_sanatci = sanatci_puanlari.sort_values(by='Valence', ascending=False).head(1)

print("En Enerjik Sanatçı:")
print(en_enerjik_sanatci)

print("\nEn Dans Edilebilir Sanatçı:")
print(en_dans_edilebilir_sanatci)

print("\nEn Duygusal Sanatçı:")
print(en_duygusal_sanatci)
```
